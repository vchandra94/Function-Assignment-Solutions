**Question-1)**  What is the difference between a function and a method in Python?

**Answer-1)** In Python, functions and methods are both used to perform actions, but they are not the same.
**Function**

1. A function is a block of reusable code that performs a specific task.

2. It is not tied to any object.

3. Defined using the def keyword.

4. Can be called independently.

Example:

```python
def add(a, b):
    return a + b

print(add(5, 3))
```

**Method:**
1. A method is a function that belongs to an object (usually an instance of a class).

2. Called using dot notation on an object.

3. Defined inside a class.

4. It automatically takes the object (self) as its first parameter.

Example:-

```python
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
print(calc.add(5, 3))
```

In short:

A function is independent.

A method is tied to an object or class.

**Question-5)** What are iterators in Python and how do they differ from iterables?

**Answer-5)**  Iterators vs. Iterables in Python
Both iterators and iterables are used to loop over data, but they are not the same.

What is an Iterable?
1. An iterable is any object you can loop over using a for loop.

2. It implements the __iter__() method.

3. Examples: list, tuple, string, set, dict, etc.

```python
my_list = [1, 2, 3]  # This is an iterable
for item in my_list:
    print(item)
```

 What is an Iterator?
An iterator is an object that produces values one at a time using the next() function.

It implements both:

__iter__() → returns itself

__next__() → returns the next value

```python
my_list = [1, 2, 3]
it = iter(my_list)

print(next(it))
print(next(it))
print(next(it))
```

**Summary:-**
1. Iterable: Can be converted into an iterator using iter().

2. Iterator: Used to fetch items one at a time using next().

**Question-6)** Explain the concept of generators in Python and how they are defined.

**Answer-6)** A generator in Python is a special type of iterator used to generate values one at a time from a function, instead of computing and storing them all at once. This makes generators efficient, especially for large datasets or infinite sequences.

**Key Features of Generators:**
1. Uses the **yield** keyword instead of return

2. Remembers its state between function calls

3. Returns a generator object (which is an iterator)

4. Values are generated on the fly (lazy evaluation)

5. Consumes less memory compared to lists

How can Define a Generator Function:
A generator is defined like a normal function but uses yield to produce a value:

```python
def count_up_to (n):
    count = 1
    while count <= n:
        yield count
        count += 1

        counter = count_up_to(3)
```

Calling this function returns a generator object:

```python
counter = count_up_to(3)

print(next(counter))
print(next(counter))
print(next(counter))
```

You can also loop through it:          |


```python
for num in count_up_to(3):
    print(num)
```

**Benefits of Generators**
1. Efficiency: Doesn’t load everything in memory.

2. Convenience: Ideal for streaming data or large files.

3. Clean Syntax: Easier to write than class-based iterators.



Example Use Case
Reading a large file line-by-line:

```python
def read_large_file(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip()
```

**Question-7)** What are the advantages of using generators over regular functions?

**Answer-7)** Advantages of Generators Over Regular Functions
Memory Efficient:
Generators yield one value at a time, so they don’t store the entire result in memory.

Faster Execution (Lazy Evaluation):
Values are generated only when needed, improving performance for large or long-running tasks.

Can Handle Infinite Data:
Generators are ideal for producing data streams or infinite sequences (like counting forever).

Simpler Code:
Easier to write than creating full iterator classes.

Better for Loops:
Great for looping through large datasets without slowing down or crashing.

In short: Generators save memory, improve speed, and simplify iteration—especially useful for large or endless data.


Example:

Regular Function (returns full list):

```python
def get_numbers():
    return [i for i in range(1000000)]

nums = get_numbers()
print(nums[0])  # Uses a lot of memory
```

 Generator Function (yields values one at a time):

```python
def get_numbers():
    for i in range(1000000):
        yield i

nums = get_numbers()
print(next(nums))  # Uses very little memory
```

Conclusion:
1. Generators are ideal when:

2. we’re working with large datasets

3. We don’t need all values at once

4. We want to save memory and speed up your program

**Question-8)** What is a lambda function in Python and when is it typically used?

**Answer-8)** A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can have any number of arguments but only one expression.

Syntax: lambda arguments: expression

Example:

```python
square = lambda x: x * x
print(square(5))
```

Lambda functions are typically used when:

1. We need a short function for a short period of time.

2. Used as an argument to higher-order functions like map(), filter(), or sorted().

3. Used in functional programming style where small, throwaway functions are helpful.

Example with map():

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
```

Key Points:
1. Lambda functions are anonymous (don't have a name unless assigned to a variable).

2. They are limited to a single expression.

3. Great for short, simple operations where defining a full function with def would be unnecessary.

**Question-9)** Explain the purpose and usage of the `map()` function in Python.

**Answer-9)** Actual Purpose and Usage of map() in Python:-

**Purpose:**
The main purpose of map() is to transform (or "map") each item in an iterable by applying a function to it — without writing a loop manually. It makes your code cleaner and often faster for element-wise operations.

**Usage:**
map(function, iterable)

1. akes two arguments:

   A function (can be built-in, user-defined, or lambda)

   An iterable (like list, tuple, etc.)

2. Returns a map object (which is an iterator), so you need to convert it using list(), tuple(), etc., to view results.


**Example 1: Using a Defined Function**

```python
def capitalize(word):
    return word.upper()

words = ['apple', 'banana', 'cherry']
result = map(capitalize, words)

print(list(result))
```

**Example 2: Using lambda Function**

```python
numbers = [1, 2, 3, 4]
squares = map(lambda x: x * x, numbers)

print(list(squares))
```

Multiple Iterables:
If the function takes multiple arguments, you can pass multiple iterables to map():

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x + y, a, b)
print(list(result))
```

**Best way to Use When:**
1. We want to apply one function to every item in an iterable.

2. You're doing simple transformations (e.g., converting, calculating, formatting).

**Conclusion:**

The map() function helps you write clean, efficient, and readable code when transforming iterables. It's especially powerful when combined with lambda for quick, in-place operations.

**Question-10)**  What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?

**Answer-10)**

Here’s a clear comparison between map(), reduce(), and filter() in Python:

**1. map()**

Purpose: Applies a function to each element in an iterable.

Returns: A new iterable (map object) with transformed values.

Syntax: map(function, iterable)

Example:-


```python
numbers = [1, 2, 3]
squares = list(map(lambda x: x * x, numbers))
print(squares)
```

**2. filter()**

Purpose: Applies a function that returns True or False to filter elements.

Returns: A new iterable (filter object) with only the items for which the function returns True.

Syntax: filter(function, iterable)

Exapmle:-

```python
numbers = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

**reduce() (from functools)**

Purpose: Applies a function to accumulate values (reduces the iterable to a single result).

Returns: A single value.

Syntax:

from functools import reduce

reduce(function, iterable)


Example:-

```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

Use Case Tips:-

Use **map()** when We're changing data.

Use **filter()** when We're choosing data.

Use **reduce()** when We're combining data into one value.

```python
#Question-4)  Write a Python function that checks if a given number is prime or not from 1 to 200.

#Answer-4)
def is_prime(num):

    if num <= 1:

        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print("Checking prime numbers from 1 to 200:")
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")
    else:
        print(f"{number} is not a prime number.")

```

```python
#Question-5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

#Answer-5)
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

n = int(input("Enter the number of Fibonacci terms to generate upto: "))
fib = FibonacciIterator(n)

print(f"\nFibonacci sequence up to {n} terms:")
for num in fib:
    print(num)
```

```python
#Question-7) Implement a generator function that reads a file line by line and yields each line as a string.

#Answer-7)
import os

def read_large_file(file_path):

    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file at '{file_path}' was not found.")

    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            yield line.rstrip('\n')


file_name = "sample.txt"
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("This is the first line.\n")
    f.write("Here is the second line.\n")
    f.write("And the third line.\n")
    f.write("Final line.")


print("Reading file using the generator:")
try:
    line_reader = read_large_file(file_name)
    for line in line_reader:
        print(line)


except (FileNotFoundError, IOError) as e:
    print(f"An error occurred: {e}")
finally:

    if os.path.exists(file_name):
        os.remove(file_name)

```

```python
#Question-10)  Create a Python program that uses 'filter()' to remove all the vowels from a given string.

#Question-10)
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, s))

# Example usage
input_str = input("Enter a string: ")
result = remove_vowels(input_str)

print("String without vowels:", result)
```
